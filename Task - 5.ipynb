```python
# Percenateg of times when the toss winner also won the match, each season
round(Final_dataset.query('toss_winner == winner').groupby('season').count()['winner']*100/Final_dataset.groupby('season').count()['winner'] ,2).reset_index(name='Percentage of when Toss winner is the Match winner')
```

INSIGHTS-    
From above data of each season on an average the chances of winning does depend on Toss winner

```python
Final_dataset['win_by_first']=np.where(Final_dataset['win_by_runs']>0,'Bat first','Bowl first')

plt.figure(figsize = (18,10))
sns.countplot(x=Final_dataset['season'],hue=Final_dataset['win_by_first'])
plt.title("Numbers of matches won by either batting or bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=10)
plt.ylabel("Count",fontsize=10)
plt.show()
```

INSIGHTS -    
From the visualisation can observe that major games are won when the team Bowls first, only 2 instances that is 2010and 2015 it wasnt the case.

```python
Toss=Final_dataset.toss_decision.value_counts()

plt.pie(Toss.values, labels=Toss.index,autopct='%.1f%%',startangle=90)
plt.title('Tendency of a Toss opting ',fontsize=20)
plt.axis('equal')
plt.show()
```

INSIGHTS -    
Observation is that After Winning The Toss, The Maximum Probability is Fielding First (61%)

```python
Final_dataset['toss_win_game_win'] = np.where((Final_dataset.toss_winner == Final_dataset.winner),'Yes','No')
plt.figure(figsize = (8,8))
sns.countplot(x=Final_dataset['toss_win_game_win'], hue = Final_dataset['toss_decision'])
plt.title("How Toss Decision Affects Match Result", fontsize = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel("Winning Toss And Winning Match", fontsize = 15)
plt.ylabel("Frequency", fontsize = 15)
plt.show()
```

INSIGHTS -    
There is somewhat equal probability of match winning or lossing when opted for batting or balling

```python
plt.figure(figsize = (15,10))
sns.countplot(x=Final_dataset['season'],hue=Final_dataset['toss_decision'])
plt.title("Matches Won By Toss Result ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

INSIGHTS -   
There is a High Observation of Winning When The Team chooses to Field First over Winning The Toss, in the years 2016-2019

```python
# which team has won max matches in which ground?
df=Final_dataset.groupby(['winner', 'venue'])['venue'].count().reset_index(name='Matches played and won')
#df.query('winner== "Chennai Super Kings"')
df['venue'] = df['venue'].replace(['Feroz Shah Kotla Ground', 'M Chinnaswamy Stadium', 'MA Chidambaram Stadium, Chepauk','Punjab Cricket Association IS Bindra Stadium, Mohali','Rajiv Gandhi International Stadium, Uppal	'], ['Feroz Shah Kotla', 'M. Chinnaswamy Stadium','M. A. Chidambaram Stadium','Punjab Cricket Association Stadium, Mohali','Rajiv Gandhi Intl. Cricket Stadium'])
df
```

```python

df.groupby(['venue'])['Matches played and won'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize = (15,10))
df.groupby(['venue'])['Matches played and won'].sum().sort_values(ascending=False)[:10].plot(kind='bar')
plt.title("Top 10 venues of maximum matches won")
plt.show()
```

INSIGHTS -   
M. Chinnaswamy Stadium has had maximum matches which were played and won

```python
# which team has won maximum times in the IPL seasons?
plt.figure(figsize=[18,5])
sns.countplot(x=Final_dataset['winner'])
plt.xticks(rotation=90)
plt.show()
```

INSIGHTS -    
Mumbai Indians has won majority of the times, followed by Chennai Super Kings and Kolkata Knight Riders respectively.

```python
Number_of_fours = Final_dataset[Final_dataset['batsman_runs'] == 4]
Number_of_fours.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')]).sort_values(by=['fours'],ascending=False)
```

INSIGHTS -    
Maximum fours in the overall IPL seasons was by Mumbai Inidans

```python
Number_of_six = Final_dataset[Final_dataset['batsman_runs'] == 6]
Number_of_six.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')]).sort_values(by=['sixes'],ascending=False)
```

INSIGHTS -     
Maximum sixes in the overall IPL seasons was by Royal Challengers Bangalore

```python
# bowling
plt.figure(figsize=(18,10))
ax=sns.countplot(x=Final_dataset['dismissal_kind'])
plt.title("Dismissals in IPL",fontsize=20)
plt.xlabel("Dismissals Kind",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

INSIGHTS -

```python
plt.figure(figsize = (20,10))
sns.countplot(x=Final_dataset['toss_winner'], hue = Final_dataset['toss_decision'])
plt.title("Teams decision on winning toss", size = 20)
plt.xticks(size = 15, rotation=90)
plt.yticks(size = 15)
plt.xlabel("Toss Winner", size = 15)
plt.ylabel("Count", size = 15)
plt.show()
```

INSIGHTS -    
Major teams decision was to always field after winning the toss. So that they could chase the score in the 2nd innings.

```python
bowler_DATA = Final_dataset.groupby('bowler').aggregate({'ball': 'count', 'total_runs': 'sum', 'player_dismissed' : 'count'})
bowler_DATA.columns = ['Runs given','Balls played','Wickets taken']
TopBowler_wickets = bowler_DATA.sort_values(['Wickets taken'], ascending=False)[:20]
TopBowler_wickets
```

```python
plt.figure(figsize = (10,10))
sns.barplot(x= TopBowler_wickets.index,y=TopBowler_wickets['Wickets taken'])
plt.title("Top Wicket takers", size = 10)
plt.xticks(size = 15, rotation=90)
plt.yticks(size = 15)
plt.show()
```

INSIGHTS -   
Top 20 players with a good bowling skills and wicket takers. At the top with Lasith Malinga, DJ Bravo, A Mishra and so on

```python
Man_of_The_Match = Final_dataset['player_of_match'].value_counts()
Man_of_The_Match.head(10).plot(kind = 'bar',figsize=(12,8), fontsize=15)
plt.title("Top 10 Players With Most Man of The Match Awards",fontsize=20)
plt.ylabel("Count", size = 15)
plt.xlabel("Players", size = 15) 
plt.show()
```

INSIGHTS -    
CH Gayle And AB de Villiers are Two Players who have won The Most Number of Man of The Match Awards In IPL.

```python
Final_dataset.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar',
                                                                                            figsize = (15,5))
plt.title("Top Run Getters of IPL", fontsize = 25)
plt.xlabel("Batsmen Name", size = 20)
plt.ylabel("Total Runs Scored", size = 20)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.show()
```

Virat Kohli And Suresh Raina Are The Top Players Who Make Top Runs In The IPL.
